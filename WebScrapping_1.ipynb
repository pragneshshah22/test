```python

# Step 1: Make a request to the website
url = 'https://www.python.org/blogs/'
response = requests.get(url)

# Step 2: Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Step 3: Find the elements that contain the article titles
titles = soup.find_all('h2', class_='widget-title')
#titles

# Step 4: Print out the titles
for tit in titles:
    print(tit.text.strip())
    
```

### soup.find_all('h2', class_='widget-title'): This line searches the parsed HTML for all <h2> tags with the class  attribute widget-title. These tags typically contain the titles of the blog articles.
### titles: A list of all matching elements (in this case, the blog post titles) is stored in the titles variable.

```python
from bs4 import BeautifulSoup
import requests

# Step 1: Make a request to the website
url = 'https://www.python.org/blogs/'
response = requests.get(url)

# Step 2: Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Step 3: Find the elements that contain the article titles
# We'll inspect the page to identify the correct tag and class
titles = soup.find_all('h3', class_='event-title')

# Step 4: Print out the titles
for title in titles:
    print(title.text.strip())
```

```python
# Check if the request was successful
response = requests.get(url)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')
    res=soup.title
    print(res.get_text())

else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

What It Does:

The response.content contains the raw HTML content of the webpage.
BeautifulSoup is used to parse this HTML content and create a structured representation of the page, making it easier to extract specific elements.
Why 'html.parser' is Used:
It specifies the parsing library to use. The built-in html.parser is lightweight and sufficient for most use cases.

```python
# Check if the request was successful
response = requests.get(url)
if response.status_code == 200:
    # Parse the HTML content of the page
    print(soup.prettify())
    soup = BeautifulSoup(response.content, 'html.parser')

else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

What It Does:
The prettify() method formats the parsed HTML into a readable and well-indented structure. This makes it easier to inspect and understand the layout of the page.

Why It’s Useful:
It helps identify the tags, IDs, or classes required for data extraction. For example, you can locate where the div with id="table_of_content" appears in the HTML.
